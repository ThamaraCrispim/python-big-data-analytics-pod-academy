## Operadores de atribuição

Em Python, operadores de atribuição são utilizados para atribuir um valor a uma variável. Eles permitem que você atualize ou modifique o valor de uma variável de forma conveniente e eficiente.

O operador mais comum é o sinal de igual (=), que é usado para atribuir um valor a uma variável. Por exemplo, na instrução "x = 5", o valor 5 é atribuído à variável "x".

+= - Atribuição com Adição \

/= - Atribuição com Divisão \
//= - Atribuição com Divisão inteira \
%= - Atribuição com Módulo \
**= - Atribuição com Potênciação

# Estruturas condicionais

Em Python, as condições são usadas para tomar decisões no código. As condições são retornam como True ou False. Essas instruções são usadas para controlar o fluxo do programa com base no resultado de uma condição específica.

## Estruturas condicionais - if

A instrução if é usada para verificar se uma determinada condição é verdadeira ou não. Se a condição for verdadeira, o bloco de código dentro da instrução if é executado. Se a condição for falsa, o bloco de código será ignorado.


```python
salario = 10000
```

```python
if salario > 5000 :
    print("O salário é maior que R$5000")

```

## Estruturas condicionais - else

A instrução if-else é usada para verificar se uma determinada condição é verdadeira ou falsa. Se a condição for verdadeira, o bloco de código dentro da instrução if é executado. Se a condição for falsa, o bloco de código dentro da instrução else é executado.


```python
salario = 4000
```

```python
if salario > 5000 :
    print("O salário é maior que R$5000")
else:
    print("O salário é menor que R$5000")
```

## Estruturas condicionais - elif

A instrução if-elif-else é usada para verificar várias condições. Se a primeira condição for verdadeira, o bloco de código dentro da instrução if é executado. Se a primeira condição for falsa, a segunda condição é verificada. Se a segunda condição for verdadeira, o bloco de código dentro da instrução elif é executado. Se nenhuma das condições for verdadeira, o bloco de código dentro da instrução else é executado.

```python
salario = 4000
```

```python
if salario > 5000 :
    print("O salário é maior que R$5000")
elif salario == 4000:
    print("O salário é R$4000")
else:
    print("O salário é menor que R$5000")
```

# Coleções

As coleções em Python são usadas para armazenar e organizar **múltiplos valores em uma única variável**. Elas são muito úteis para lidar com conjuntos de dados e tornam o código mais fácil de ler e escrever.

Quando falamos de coleções, temos 4 opções: **Listas, Dicionários , Sets e Tuplas**

Alguns pontos que ressaltam a importância do uso de coleções :

Acesso eficiente aos elementos: As coleções permitem que você acesse facilmente os elementos individuais que estão armazenados dentro delas. Por exemplo, em uma lista, você pode acessar um elemento pelo seu índice.

Modificação de dados: Em algumas coleções, como as listas e dicionários, você pode modificar os dados armazenados dentro delas. Isso permite que você atualize facilmente valores existentes ou adicione novos valores.

Facilidade de iteração: As coleções em Python facilitam a iteração sobre seus elementos usando loops. Isso permite que você processe facilmente cada valor armazenado em uma coleção.

Armazenamento de valores relacionados: As coleções são úteis para armazenar valores relacionados entre si. Por exemplo, um dicionário pode ser usado para armazenar informações sobre uma pessoa, como nome, idade e endereço.

## Estruturas de repetição - break

"break" é uma palavra-chave em Python que é usada para interromper a execução de um loop (como "while" ou "for") antes que a condição de teste do loop seja avaliada como falsa. Quando o Python encontra o comando "break" dentro do corpo do loop, ele interrompe a execução do loop imediatamente e continua a execução do código que vem depois do loop.



```python
num  = 10
while True:
    if num == 10:
        print('num >', num)
        break


```

## Estruturas de repetição - continue

"continue" é outra palavra-chave em Python que é usada dentro de loops para pular a execução do código restante dentro do loop para a próxima iteração. Quando o Python encontra o comando "continue" dentro do corpo do loop, ele interrompe a execução do código restante dentro do loop para a iteração atual e passa para a próxima iteração.



```python
num = 0


while num < 10:



    num+= 1

    if num == 4:
        continue

    print(num)

```

# O comando pass

Quando a palavra-chave "pass" é usada em um bloco de código, ela não faz nada e não causa nenhum erro. "pass" normalmente é usado como um espaço reservado para um bloco de código que ainda não foi implementado.




```python
num = 0


while num < 10:



    num+= 1

    if num == 4:
        pass

    print(num)
```

# List Comprehension

List comprehension em Python é uma maneira concisa e elegante de criar novas listas a partir de sequências existentes (listas, tuplas, dicionários, etc.) usando uma sintaxe compacta e legível.

Sintaxe :

**[x for x in sequencia ]** > Básico, percorre a lista e mostra os valores.

**[expressao for x in sequencia ]** > transforma cada item na sequencia com o valor contido na expressao

**[expressao for x in sequencia if condicao ]** > transforma cada item da sequencia só que agora com uma condição

**[expressao for x in sequencia if condicao and/or condicao ]** > transforma cada item da sequencia só que agora com múltiplas condições

para usar a cláusula else, é preciso usar outra sintaxe com o operador condicional ( Ternary operator )

**[expressao if condicao else expressao for x in sequencia ]**


```python
lista1 = [1,2,3]
```

```python
#Uso básico , só percorrendo os elementos da lista1

[ num for num in lista1 ]
```

```python
#Uso básico , alterando a lista1 (cada elemento da lista1 * 2)

[ num * 2 for num in lista1 ]
```

```python
lista1
```

```python
#Agora com uma condição
[ num * 2 for num in lista1 if num == 2 ]
```

```python
#Agora com múltiplas condições
[ num * 2 for num in lista1 if num == 2 or num == 3 ]
```

```python
#usando a cláusula else
[num*2 if num == 1 else num for num in lista1 ]
```

```python
print("CADASTRO FUNCIONARIOS")
print("1 - Cadastrar \n 2 - Excluir \n 3 - Listar \n 4- sair \n")

opcao = int(input("Digite a opção desejada > "))

while opcao != 4:

    if opcao == 1:
        lista_func.append(input("Digite o nome do funcionário > "))
    elif opcao == 2:
        lista_func.remove(input("Digite o nome do funcionário a excluir > "))
    elif opcao == 3:
        print(lista_func)

    opcao = int(input("Digite a opção desejada > "))
```
