## Tipos de dados - Numéricos

Em Python, existem três tipos principais de **dados numéricos**:

**Números inteiros (int)**: Inteiros são números inteiros, positivos e negativos, sem qualquer ponto decimal. Por exemplo, 10, -7 e 0 são todos números inteiros em Python. A função int() pode ser usada para converter outros tipos de dados em números inteiros.

**Números decimais (float)**: Números de ponto flutuante, ou floats, são números reais com decimais. Por exemplo, 3.14, -2.5 e 0.0 são todos flutuantes em Python. A função float() pode ser usada para converter outros tipos de dados em floats.

**Números Complexos (complex)**: Números complexos são números com uma parte real e uma parte imaginária, representados em Python pelo sufixo j. Por exemplo, 2+3j, -1-2j e 0+1j são todos números complexos em Python. A função complex() pode ser usada para criar números complexos de outros tipos de dados.

## Operadores Aritméticos

Em Python, existem vários operadores aritméticos usados para realizar operações matemáticas em tipos de dados numéricos.

**Adição (+)**: Soma dois valores juntos. Por exemplo, 2 + 3 retorna 5.

**Subtração (-)**: Subtrai um valor de outro. Por exemplo, 5 - 2 retorna 3.

**Multiplicação (*)**: Multiplica dois valores juntos. Por exemplo, 2 * 3 retorna 6.

**Divisão (/)**: Divide um valor por outro. Por exemplo, 6 / 3 retorna 2,0 (observe que o resultado é um float, mesmo que ambos os operandos sejam inteiros).

**Floor Division (//)**: Divide um valor por outro, descartando qualquer resto e retornando apenas o quociente como inteiro. Por exemplo, 7 // 3 retorna 2.

**Módulo (%)**: Divide um valor por outro e retorna o resto. Por exemplo, 7% 3 retorna 1.

**Exponenciação (****) *: Eleva um valor à potência de outro. Por exemplo, 2 ** 3 retorna 8 (ou seja, 2 elevado à potência de 3).

## Deu erro! E agora ?

Em Python, um **traceback** é um relatório que mostra a sequência de chamadas de função que levaram a um erro. Quando ocorre um erro em um programa Python, o Python normalmente imprime um rastreamento no console, que mostra a linha de código que causou o erro, bem como a sequência de chamadas de função que levaram a essa linha de código.

Tracebacks podem ser úteis para depurar o código Python, pois mostram a sequência exata de chamadas de função que levaram a um erro. Ao examinar o traceback, muitas vezes você pode identificar a causa raiz do erro e fazer as alterações necessárias para corrigi-lo.

## Extensões de arquivos ( .ipynb x .py )

Os arquivos .ipynb, também conhecidos como arquivos Jupyter Notebook, são documentos interativos que permitem criar e compartilhar código, visualizações e explicações de texto em um único documento. Eles são uma combinação de células de código e células de remarcação. As células de código permitem escrever e executar código, enquanto as células de marcação permitem adicionar texto, imagens e títulos formatados. Eles são comumente usados em fluxos de trabalho de ciência de dados e aprendizado de máquina para análise exploratória, documentação e compartilhamento de resultados.

Por outro lado, os arquivos .py são arquivos de script Python que contêm apenas código Python. Eles são normalmente usados para escrever programas autônomos ou módulos que podem ser executados a partir da linha de comando ou importados para outros programas Python. Eles não contêm nenhuma documentação ou visualização e não são interativos.
