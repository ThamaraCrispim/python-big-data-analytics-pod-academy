# Breve timeline da história dos Videogames

**Década de 1950**:

1952: O jogo "OXO" (ou "Noughts and Crosses") é criado pelo cientista da computação A.S. Douglas no Reino Unido, considerado um dos primeiros jogos digitais da história.
1958: O jogo "Tennis for Two" é criado por William Higinbotham nos Estados Unidos, utilizando um osciloscópio.
Década de 1960:

1962: O jogo "Spacewar!" é criado por estudantes do Massachusetts Institute of Technology (MIT) nos Estados Unidos.
1967: A empresa alemã Rainer Baumann desenvolve a "The Brown Box", que se tornaria o primeiro protótipo do que viria a ser o console de videogame Magnavox Odyssey.
Década de 1970:

1972: O Magnavox Odyssey é lançado nos Estados Unidos, sendo o primeiro console de videogame comercialmente disponível.
1975: A Atari é fundada nos Estados Unidos por Nolan Bushnell e Ted Dabney, lançando seu primeiro jogo, "Pong".
1977: A Atari lança o Atari 2600, um dos consoles mais populares e influentes da história.
1978: A empresa japonesa Nintendo lança seu primeiro console, o Color TV-Game, no mercado doméstico japonês.
Década de 1980:

1983: A crise dos videogames ocorre nos Estados Unidos, devido a uma saturação do mercado e baixa qualidade de muitos jogos lançados na época.
1985: A **Nintendo** lança o NES (Nintendo Entertainment System), que se tornaria um enorme sucesso global.
1989: A Sega lança o Mega Drive (conhecido como Genesis nos Estados Unidos), que se tornaria um dos principais concorrentes do NES.

**Década de 1990**:

1990: A **Nintendo** lança o Game Boy, um console portátil que se tornaria um dos mais populares da história.
1994: A **Sony** lança o PlayStation, que se tornaria um grande sucesso e estabeleceria a empresa como uma grande força no mercado de videogames.
1995: A Sega lança o Saturn, mas ele não consegue competir com o PlayStation e o Nintendo 64, que são lançados no mesmo ano.
1998: A Valve Corporation lança o jogo "Half-Life", que se tornaria um dos mais influentes e aclamados da história.
Década de 2000:

2000: A **Sony** lança o PlayStation 2, que se tornaria o console mais vendido da história, com mais de 155 milhões de unidades vendidas.
2001: A **Microsoft** lança o Xbox, entrando no mercado de consoles pela primeira vez.
2004: A **Nintendo** lança o Nintendo DS, um console portátil com tela sensível ao toque e duas telas, que se tornaria um grande sucesso.
2005: A **Microsoft** lança o Xbox 360, que se tornaria um concorrente forte do PlayStation 3 da Sony.

**Década de 2010**:

2012: A Ouya é lançada no mercado, sendo uma das primeiras iniciativas de console de jogos Android.
2013: A **Sony** lança o PlayStation 4, que se tornaria um enorme sucesso, com mais de 110 milhões de unidades vendidas em todo o mundo.
2017: A **Nintendo** lança o Switch, um console híbrido que pode ser usado tanto como um console de mesa quanto como um console portátil, que se tornaria um dos grandes sucessos da empresa.
2019: A Google lança o Stadia, um serviço de streaming de jogos que não requer hardware de console.
Década de 2020:

2020: A **Sony** lança o PlayStation 5 e a **Microsoft** lança o Xbox Series X e Series S, dando início à nova geração de consoles.
2021: A **Nintendo** lança o Switch OLED, uma versão atualizada do Switch com tela OLED.
Atualmente, há uma ampla variedade de consoles e plataformas de jogos, incluindo jogos para celular, PCs, consoles de mesa e dispositivos de streaming.
Essa é apenas uma breve timeline da história dos videogames, com algumas das principais marcas e eventos marcantes. Obviamente, há muitos outros acontecimentos, empresas e jogos que também foram importantes para a evolução dos videogames.

Atualmente, as três principais empresas que lideram o mercado de videogames são a **Sony**, a **Microsoft** e a **Nintendo**.

# EDA ( Exploratory Data Analysis)

```python
dados_games.head()
```

```python
# vamos ver a atualização desse dataset

print('Ano mais recente > ',dados_games['Year'].max())
print('Ano mais antigo > ',dados_games['Year'].min())
```

```python
#Verificando quantidade de nulos

dados_games.isnull().sum()
```

Podemos notar que existem alguns jogos que não tem o ano (year)  cadastrado, e também alguns que estão sem a Publisher.

```python
# Jogos mais vendidos
dados_games.sort_values('Global_Sales', ascending=False).head(10)
```

```python
#Qual é a relação entre o gênero dos jogos e as vendas globais?


# Agrupando as vendas globais por gênero
genre_sales = dados_games.groupby('Genre')['Global_Sales'].sum().reset_index().sort_values('Global_Sales',ascending=False)
genre_sales

# Plotando um gráfico de barras das vendas globais por gênero
plt.bar(genre_sales['Genre'] ,genre_sales['Global_Sales'] ,edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

```python
#Qual é a relação entre o plataforma dos jogos e as vendas globais?



# Agrupando as vendas globais por gênero
platform_sales = dados_games.groupby('Platform')['Global_Sales'].sum().reset_index().sort_values('Global_Sales',ascending=False)
platform_sales = platform_sales[0:15]

# Plotando um gráfico de barras das vendas globais por gênero
plt.bar(platform_sales['Platform'] ,platform_sales['Global_Sales'] ,color = '#7B68EE', edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

```python
# Criando um DataFrame apenas com as colunas de vendas por região e ano
df_sales_region_year = dados_games[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Year']]

# Agrupando as vendas por ano e região
sales_by_region_year = df_sales_region_year.groupby(['Year']).sum()

# Plotando um gráfico de linhas para as vendas por região
plt.plot(sales_by_region_year.index, sales_by_region_year['NA_Sales'], label='América do Norte')
plt.plot(sales_by_region_year.index, sales_by_region_year['EU_Sales'], label='Europa')
plt.plot(sales_by_region_year.index, sales_by_region_year['JP_Sales'], label='Japão')
plt.plot(sales_by_region_year.index, sales_by_region_year['Other_Sales'], label='Outras Regiões')
#plt.plot(sales_by_region_year.index, sales_by_region_year['Global_Sales'], label='Vendas Globais')
plt.title('Vendas de games por região 1980 - 2020')
plt.xlabel('Ano')
plt.ylabel('Vendas (milhões de dólares)')
plt.legend()
plt.show()
```

```python
#Qual é a relação entre as publishers dos jogos e as vendas globais?



publisher_sales = dados_games.groupby('Publisher')['Global_Sales'].sum().reset_index().sort_values('Global_Sales',ascending=False)
publisher_sales = publisher_sales[0:15]

# Plotando um gráfico de barras das vendas globais por publisher
plt.bar(publisher_sales['Publisher'] ,publisher_sales['Global_Sales'] ,color = '#3A68EE', edgecolor='black')
plt.xticks(rotation=90)
plt.show()

```

```python
#Qual é a relação entre o ano de lançamento e as vendas globais de jogos?

# Agrupando as vendas por ano
sales_by_year = dados_games.groupby(['Year']).sum(numeric_only=True)

# Plotando um gráfico de linhas das vendas globais por ano
plt.plot(sales_by_year.index, sales_by_year['Global_Sales'])
plt.title('Vendas globais de jogos por ano')
plt.xlabel('Ano')
plt.ylabel('Vendas globais (milhões de dólares)')
plt.show()
```

```python
# Selecionando as colunas relevantes
df_sales_region = dados_games[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

# Agrupando as vendas por região
sales_by_region = df_sales_region.sum()

plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%')
plt.title('Share de vendas por região')
plt.show()
```

```python
#quantidade de jogos por Ano

dados_games.value_counts('Year')
```

```python
#visualizando quais as plataformas dos jogos onde o ano de lançamento não foi preenchido

dados_nulos_year = dados_games[dados_games['Year'].isnull()]

dados_nulos_year.groupby('Platform').agg({'Global_Sales':'sum'}).sort_values('Global_Sales',ascending=False)
```
