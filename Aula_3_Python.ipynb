# Coleções

As coleções em Python são usadas para armazenar e organizar **múltiplos valores em uma única variável**. Elas são muito úteis para lidar com conjuntos de dados e tornam o código mais fácil de ler e escrever.

Quando falamos de coleções, temos 4 opções: **Listas, Dicionários , Sets e Tuplas**

Alguns pontos que ressaltam a importância do uso de coleções :

Acesso eficiente aos elementos: As coleções permitem que você acesse facilmente os elementos individuais que estão armazenados dentro delas. Por exemplo, em uma lista, você pode acessar um elemento pelo seu índice.

Modificação de dados: Em algumas coleções, como as listas e dicionários, você pode modificar os dados armazenados dentro delas. Isso permite que você atualize facilmente valores existentes ou adicione novos valores.

Facilidade de iteração: As coleções em Python facilitam a iteração sobre seus elementos usando loops. Isso permite que você processe facilmente cada valor armazenado em uma coleção.

Armazenamento de valores relacionados: As coleções são úteis para armazenar valores relacionados entre si. Por exemplo, um dicionário pode ser usado para armazenar informações sobre uma pessoa, como nome, idade e endereço.

### Função que calcula a similaridade entre dois textos

```python
from difflib import SequenceMatcher

def similaridade_texto(texto1, texto2):
    # Cria um objeto SequenceMatcher com as strings de entrada
    seq = SequenceMatcher(None, texto1, texto2)

    # Retorna a similaridade como um número entre 0 e 1
    return seq.ratio()
```

```python
similaridade_texto('Felipe','Jardim')
```

### Função que envia e-mail

```python
# bibliotecas necessárias
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import smtplib


USER = 'EMAIL_ENVIO'
PASSWORD = 'SENHA_EMAIL_ENVIO'
SMTP_SERVER = 'smtp.gmail.com:587'


def envia_email(assunto,destinatario):

    html = '''
            <html>
            <head>
            <body>
            <p style="font-size: 32px; border-collapse: collapse;font-family: Tahoma, Geneva, sans-serif; color: #636363">Olá! </h1>
            <p style="font-size: 30px; border-collapse: collapse;font-family: Tahoma, Geneva, sans-serif; color: #636363">Aviso de e-mail da Pod Academy! Vá estudar! :D </p>
            </body>
            </html>
            '''


    message = MIMEMultipart(
        'alternative', None, [MIMEText(html,'html')])

    message['From'] = USER
    message['To'] = destinatario
    message['Subject'] = assunto

    server = smtplib.SMTP(SMTP_SERVER)
    server.ehlo()
    server.starttls()
    server.login(USER, PASSWORD)
    server.sendmail(USER, destinatario, message.as_string())
    server.quit()

```

```python
envia_email('assunto','destinatario')
```

### Função que verifica se o documento é CPF ou CNPJ

```python
import re

def valida_cpf_cnpj(doc):
    if not doc:  # verifica se o documento foi informado
        return False

    doc = re.sub('[^0-9]', '', str(doc))  # remove caracteres não numéricos

    # CPF
    if len(doc) == 11:
        if doc == '00000000000' or doc == '11111111111' or doc == '22222222222' or doc == '33333333333' or doc == '44444444444' or doc == '55555555555' or doc == '66666666666' or doc == '77777777777' or doc == '88888888888' or doc == '99999999999':
            return False

        # cálculo do primeiro dígito verificador
        soma = 0
        for i in range(9):
            soma += int(doc[i]) * (10 - i)
        resto = 11 - (soma % 11)
        if resto == 10 or resto == 11:
            resto = 0
        if resto != int(doc[9]):
            return False

        # cálculo do segundo dígito verificador
        soma = 0
        for i in range(10):
            soma += int(doc[i]) * (11 - i)
        resto = 11 - (soma % 11)
        if resto == 10 or resto == 11:
            resto = 0
        if resto != int(doc[10]):
            return False

        return 'CPF'

    # CNPJ
    elif len(doc) == 14:
        if doc == '00000000000000' or doc == '11111111111111' or doc == '22222222222222' or doc == '33333333333333' or doc == '44444444444444' or doc == '55555555555555' or doc == '66666666666666' or doc == '77777777777777' or doc == '88888888888888' or doc == '99999999999999':
            return False

        # cálculo do primeiro dígito verificador
        soma = 0
        pesos = [5, 4, 3, 2, 9, 8, 7, 6, 5, 4, 3, 2]
        for i in range(12):
            soma += int(doc[i]) * pesos[i]
        resto = 11 - (soma % 11)
        if resto == 10 or resto == 11:
            resto = 0
        if resto != int(doc[12]):
            return False

        # cálculo do segundo dígito verificador
        soma = 0
        pesos = [6, 5, 4, 3, 2, 9, 8, 7, 6, 5, 4, 3, 2]
        for i in range(13):
            soma += int(doc[i]) * pesos[i]
        resto = 11 - (soma % 11)
        if resto == 10 or resto == 11:
            resto = 0
        if resto != int(doc[13]):
            return False

        return 'CNPJ'
    else:
        return False
```

```python
valida_cpf_cnpj('00000000000456')
```

# Visibilidade de variáveis

Em python, temos "camadas" de visibilidade de variáveis, que determinam as áreas do código onde aquela variável pode ser acessada.
Chamamos essas "camadas" de escopos, e temos três tipos diferentes : **Global , Local e Nonlocal**.

## Escopo global

Em Python, uma variável global é uma variável que pode ser acessada em qualquer parte do programa, incluindo dentro de funções ou métodos. Variáveis globais são definidas fora de funções ou métodos e não são limitadas ao escopo de uma função específica.

Para definir uma variável global em Python, basta declará-la fora de qualquer função ou método.

Para acessar uma variável global dentro de uma função, precisamos colocar a palavra "global" antes da variável.

```python
x = 10 # variável global
```

```python
x
```

## Escopo local

Variáveis locais são aquelas que só podem ser acessadas dentro da função ou método em que foram definidas. Quando uma variável é definida dentro de uma função ou método, ela é automaticamente considerada como uma variável local.

```python
x = 10 # variável global

def exibe_numero():
    global x

    return x

```

```python
def exibe_numero():
    y = 10

    return y
```

```python
# a variável não é global, só funciona dentro da função.
y
```

## Escopo Nonlocal

Variáveis "nonlocal" são aquelas que podem ser acessadas dentro de funções aninhadas (nested functions). Elas são usadas para acessar variáveis de uma função externa à função atual. Para definir uma variável nonlocal, ela deve ser declarada dentro da função aninhada e ser precedida pela palavra-chave "nonlocal".

```python
def funcao_teste():
    x = 5  # variável local

    def funcao_aninhada():
        nonlocal x
        x = 10
        print(x)

    funcao_aninhada()
    print(x)

funcao_teste()
```

## _ _ name _ _  == ' _ _ main _ _ ' ??

Em Python, a expressão _ _name_ _ == "_ _main_ _" é comumente usada para verificar se um script Python está sendo executado como um programa principal ou se está sendo importado como um módulo em outro script.

Quando um script Python é executado, a variável name é definida como "main". Se o script estiver sendo importado como um módulo em outro script Python, a variável name receberá o nome do módulo em vez de "main". Isso permite que você diferencie a execução de um script como um programa principal ou como um módulo em outro programa.

Portanto, ao usar a expressão "name == main", você pode escrever código que será executado somente se o script estiver sendo executado como um programa principal e não será executado se o script estiver sendo importado como um módulo em outro programa. Isso é útil para escrever código de inicialização ou testes que devem ser executados apenas quando o script é executado como um programa principal.

# Estrutura Match Case

"Match case" é uma funcionalidade adicionada ao **Python na versão 3.10** que permite realizar padrões de correspondência em expressões condicionais.

Essa funcionalidade pode ser usada com a instrução match, que é uma alternativa mais poderosa à instrução switch que é encontrada em outras linguagens de programação.

## Exemplos

```python
def codigos_http(codigo):

    match codigo:
        case 100:
            print("Informações")
        case 200:
            print("Sucesso!!!")
        case 400:
            print("Erro de acesso")
        case 500:
            print("erro no servidor")
        case _:
            print("Erro desconhecido")
```

```python
codigos_http(600)
```

# Tratamento de exceções

É uma técnica para lidar com erros que ocorrem durante a execução do programa. Quando um erro é encontrado, o Python lança uma exceção que pode ser tratada por meio de um bloco de código específico. O tratamento de exceções permite que o programa controle o erro de forma mais elegante e evite que o programa seja interrompido abruptamente.

Em Python, um bloco de código para tratamento de exceções é definido usando as palavras-chave **try, except e finally**. O bloco try contém o código que pode gerar exceções, e o bloco except contém o código que será executado se a exceção ocorrer. O bloco finally é opcional e contém o código que será sempre executado, independentemente de haver ou não uma exceção.

## Abrir arquivos

A função open() é utilizada em Python para abrir arquivos em diferentes modos (leitura, escrita, etc..). A sintaxe básica da função open() é a seguinte:

open(nome_do_arquivo, modo, encoding)


nome_do_arquivo: é o caminho completo ou relativo para o arquivo que você deseja abrir.

modo: é o modo em que o arquivo deve ser aberto, como leitura ('r'), escrita ('w'), ou adição ('a').

encoding (opcional): é a codificação do arquivo que você está abrindo. Se não for especificado, a codificação padrão do sistema será utilizada.

## Ler arquivos

```python
f = open("exemplo.txt", "r")
print(f.read())
```

## Escrever arquivos

Para escrever em um arquivo existente, você precisa adicionar parâmetros na função open():

"a" - Append - Vai escrever dados no final do arquivo.

"w" - Write - Vai sobreescrever o arquivo.

```python
f = open("exemplo.txt", "w")
f.write("\nteste\nteste")
f.close()

#open and read the file after the appending:
f = open("exemplo.txt", "r")
print(f.read())
```

```python
f = open("exemplo.txt", "w")
f.write("OPA, APAGUEI TUDO ")
f.close()

#open and read the file after the overwriting:
f = open("exemplo.txt", "r")
print(f.read())
```

## Usando a expressão WITH

A declaração with é usada para criar um contexto em que um objeto é usado, garantindo que os recursos associados a esse objeto sejam liberados no final do contexto. Isso é útil para garantir que os recursos, como arquivos abertos, sejam fechados adequadamente, evitando vazamentos de recursos e problemas de desempenho.

```python
f = open("exemplo.txt", "r")
print(f.read())
f.close()
```

```python
with open("exemplo.txt", "r") as arquivo:
    conteudo = arquivo.read()
```

# Ambientes virtuais (Venvs)

É muito importante entendermos que, ao instalar o Python, temos como configurar Ambientes Virtuais ( Venvs ) que são muito úteis no desenvolvimento de aplicações e projetos.

Venvs são ambientes virtuais em python, onde podemos separar os pacotes e configurações de cada desenvolvimento/projeto. Sua principal função é manter os projetos íntegros.

Vamos imaginar dois cenários para entender um possível problema de não separar os projetos em ambientes diferentes:

 Cenário 1 :

Imagina que você precisa desenvolver um Software capaz de criar uma calculadora simples em python e pra esse projeto, precisa usar um pacote chamado numpy.

 Cenário 2:

Imagine que você precisa desenvolver um Software muito específico de Calculadora Científica em Python e ele precisa sempre ficar atualizado. Nesse cenário, também usamos numpy.



Agora temos dois projetos rodando, muito bom !


 Problema:

Depois de meses, surgiu uma atualização no pacote numpy. Ao atualizar o pacote, se não houver nenhuma Venv configurada, o pacote é atualizado no ambiente geral, que contempla todos projetos.
Nossa calculadora científica funcionou, ficou legal com essa nova atualização!
Nossa calculadora simples que está com o código bem desatualizado, parou de funcionar! Ops!

## Repositórios e gerenciadores de pacotes

### Package Installer for Python - PIP

PIP (abreviação de "pip installs packages") é um sistema de gerenciamento de pacotes para Python usado para instalar, atualizar e gerenciar pacotes Python e suas dependências. É o instalador de pacote padrão para Python e vem junto com a maioria das instalações do Python.

O PIP funciona conectando-se ao PyPI (Python Package Index) para baixar e instalar pacotes. Ele instala automaticamente todas as dependências exigidas por um pacote, garantindo que todos os pacotes necessários estejam disponíveis e atualizados. O PIP também permite que os desenvolvedores especifiquem qual versão de um pacote desejam instalar, facilitando o gerenciamento de versões e dependências do pacote.

Usando o PIP, os desenvolvedores podem facilmente instalar pacotes a partir da linha de comando digitando "pip install nome-do-pacote". O PIP também fornece opções para atualizar, desinstalar e listar pacotes instalados.

Além disso, o PIP suporta o uso de arquivos de requisitos, que permitem aos desenvolvedores especificar uma lista de pacotes necessários e suas versões em um arquivo de texto. Isso facilita o compartilhamento e a reprodução das dependências de um projeto em diferentes ambientes.

No geral, o PIP é uma ferramenta essencial para desenvolvedores Python, facilitando o gerenciamento de pacotes Python e suas dependências, garantindo que seu código esteja sempre usando as versões corretas dos pacotes necessários.

### Anaconda Repo

O Anaconda Repository é um serviço oferecido pela empresa Anaconda que permite aos usuários hospedar e compartilhar pacotes de software Python, bem como outros tipos de arquivos binários, como R, Java, C ++ e outros, em um repositório centralizado.

O Anaconda Repository é usado principalmente pelos usuários do Anaconda Distribution, que é uma distribuição popular de Python e outras bibliotecas científicas de código aberto para análise de dados, aprendizado de máquina e ciência de dados. Ele permite que os usuários compartilhem seus próprios pacotes e bibliotecas com a comunidade de usuários do Anaconda, além de permitir que os usuários acessem pacotes e bibliotecas desenvolvidos por outras pessoas.

Os pacotes hospedados no Anaconda Repository podem ser facilmente instalados e gerenciados usando o Anaconda Navigator ou o comando "conda" em um terminal. Além disso, o Anaconda Repository oferece recursos adicionais, como controle de versão, permissões de acesso e monitoramento de uso.

## requirements.txt

O arquivo "requirements.txt" é um arquivo de texto simples que contém uma lista de dependências necessárias para um projeto em Python. Cada linha do arquivo especifica o nome de um pacote ou biblioteca e sua versão correspondente, separados por um sinal de igual ("=").

O objetivo principal do arquivo "requirements.txt" é permitir que outros desenvolvedores possam facilmente reproduzir o ambiente de desenvolvimento em que o projeto foi construído, garantindo que todas as dependências necessárias estejam instaladas e configuradas corretamente.

O arquivo "requirements.txt" é comumente utilizado em conjunto com ferramentas de gerenciamento de pacotes do Python, como o pip, que podem ler o arquivo e instalar automaticamente todas as dependências listadas.
